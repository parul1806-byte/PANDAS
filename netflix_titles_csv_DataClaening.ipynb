## Clean Missing Values

```python
file['country'] = file['country'].fillna("unknown")
file['cast'] = file['cast'].fillna("not given")
file['director'] = file['director'].fillna("not given")
```

## Clean Whitespace & Text

```python
file['type'] = file['type'].str.strip()
file['rating'] = file['rating'].str.strip()
```

## Convert date_added to datetime

```python
file['date_added'] = file['date_added'].astype("string").str.strip()
file['date_added'] = pd.to_datetime(file['date_added'])
file['year_added'] = file['date_added'].dt.year
file['year_added'] = file['year_added'].astype('Int64')
file['month_added'] = file['date_added'].dt.month
file['month_added'] = file['month_added'].astype('Int64')
```

## Remove Duplicates

```python
file = file.drop_duplicates(subset = "title",keep = "first")
```

```python
file.to_csv("netflix_cleaned1.csv",index = False)
```
