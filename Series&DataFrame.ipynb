## Arithmetic Operations

```python
df = {"product":["A","B","C"],"price":[30,60,56],"Quantity":[10,5,15]}
df1 = pd.DataFrame(df)
print(df1)
```

```python
#Add a new column: Total = Price * Quantity
df1['total'] = df1['price']*df1['Quantity']
print((df1['total']))
print()
print(df1)
```

```python
#Increase Price by 10% and store in a new column NewPrice
df1['newprice'] = df1['price']+((df1['price']*10)/100)
print(df1)
```

```python
#Compare: NewPrice - Price
df1['compareprice'] = df1['newprice'] - df1['price']
print(df1)
```

```python
#Create a DataFrame of 5 countries with columns:Country,Population (millions),Area (km²),Continent:
countrydata = {"country":["India","China","America","Italy","Turkey"],
            "Population(millions)":[1428,1412,339,59,85],
            "Area(km^2)":[3287263,9596961,9833520,301340,783562],
            "Continent":['Asia','Asia','NorthAmerica','Europe','Europe']}
dataframe = pd.DataFrame(countrydata)
print(dataframe)
```

```python
#Compute Population Density
dataframe['Population_Density'] = dataframe['Population(millions)']*1000000000/dataframe['Area(km^2)']
print(dataframe['Population_Density'])
```

```python
#Show only Asian countries
print(dataframe[dataframe['Continent']=='Asia'])
```

```python
#Sort by density descending
print(dataframe.sort_values(by='Population_Density',ascending = False))
```

## Conditional Column Creation

```python
#Create a new column "result" that contains "Pass" if score ≥ 60, else "Fail".
df = pd.DataFrame({
    "name": ["A", "B", "C", "D"],
    "score": [55, 92, 67, 40]
})
result = []
for i in df['score']:
    if i >= 60:
        result.append('pass')
    else:
        result.append('fail')
df['result'] = result
print(df)
        
        
    
```

## Basic DataFrame Arithmetic

```python
#Add a new column "total" = math + science
df = pd.DataFrame({
    "math": [60, 70, 80],
    "science": [65, 75, 85]
})
df['total'] = df['math']+df['science']
df
```

```python
#Add a new column "average" = total / 2
df['average'] = df['total']/2
df
```

```python
#Create a Series containing the sum of each row.
df = pd.DataFrame({
    "a": [1, 4, 7],
    "b": [2, 5, 8],
    "c": [3, 6, 9]
})
df.sum(axis = 0)
```

## Replace Values Using Arithmetic

```python
#Replace all negative values with their positive version (absolute value) without using abs().

s = pd.Series([3, -2, 10, -5, 8])
s[s<0] = s[s<0]*-1
s
```

## Combine Series Into a DataFrame

```python
#Create a DataFrame with columns "name" and "mk".
names = pd.Series(["A", "B", "C"])
marks = pd.Series([80, 95, 88])
df= pd.DataFrame({"name":names,"mk":marks})
print(df)
```

```python
#Create a column "normalized" where each value = x / max(x)
df = pd.DataFrame({
    "x": [10, 20, 30, 40]
})
max_val = np.max(df['x'])
df['normalized'] = df['x']/max_val
df
```

## Quick Missing Value Check

```python
#Create a new column "missing_count" containing number of NaNs in each row.
df = pd.DataFrame({
    "col1": [1, None, 3],
    "col2": [4, 5, None]
})

print(df)
df['missing_value'] = df.isna().sum(axis = 1)
df

    
```
